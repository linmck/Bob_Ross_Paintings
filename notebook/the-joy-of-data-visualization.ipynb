```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ast import literal_eval
```

```python
data_filepath = os.path.join('..', 'data', 'bob_ross_paintings.csv')
joy_of_painting_df = pd.read_csv(data_filepath, index_col=0)
joy_of_painting_df.head()
```

```python
color_hex = []
for row in range(len(joy_of_painting_df)) : 
    number_of_colors = joy_of_painting_df.iloc[row,5]
    color_name = literal_eval(joy_of_painting_df.iloc[row, 7])
    hex_values = literal_eval(joy_of_painting_df.iloc[row, 8])
    color_and_hex = {color_name[i]: hex_values[i] for i in range(number_of_colors)} 
    color_hex.append(color_and_hex)
    
color_hex_dic = []
for dic in color_hex:
    for val,key in dic.items():
        color_hex_value = val,key
        color_hex_dic.append(color_hex_value)
        
unique_color_list = list(set(color_hex_dic))
print(unique_color_list)
```

```python
color_df = pd.DataFrame(unique_color_list, columns =['name', 'hex']) 
```
